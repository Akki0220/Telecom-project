```python
# Filling numerical missing values with meadian
numerical_cols = df.select_dtypes(include=['float64','int64']).columns
df[numerical_cols] = df[numerical_cols].apply(lambda x: x.fillna(0))
```

```python
# Identify and treat outliers (using z-score)
from scipy import stats

z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]

# Display updated data
print(df.describe())
```

```python
# Renaming columns for better understanding
df.rename(columns={
    'Bearer Id': 'Number_of_xDR_Sessions',
    'Dur. (ms)': 'Session_Duration',
    'Total DL (Bytes)': 'Total_DL',
    'Total UL (Bytes)': 'Total_UL',
    'Social Media DL (Bytes)':'Social_Media_DL',
    'Social Media UL (Bytes)': 'Social_Media_UL',
    'Google DL (Bytes)': 'Google_DL',
    'Google UL (Bytes)': 'Google_UL',
    'Email DL (Bytes)': 'Email_DL',
    'Email UL (Bytes)': 'Email_UL',
    'Youtube DL (Bytes)': 'Youtube_DL',
    'Youtube UL (Bytes)': 'Youtube_UL',
    'Netflix DL (Bytes)': 'Netflix_DL',
    'Netflix UL (Bytes)': 'Netflix_UL',
    'Gaming DL (Bytes)': 'Gaming_DL',
    'Gaming UL (Bytes)': 'Gaming_UL',
    'Other DL (Bytes)': 'Other_DL',
    'Other UL (Bytes)': 'Other_UL'
}, inplace=True)
```

```python
# Making a short interpretation
top_users_by_session = df.sort_values(by='Number_of_xDR_Sessions', ascending=False).head(10)
top_users_by_data = df.sort_values(by=['Total_DL', 'Total_UL'], ascending=False).head(10)

print("Top 10 users by number of sessions:")
print(top_users_by_session)

print("\nTop 10 users by data usage:")
print(top_users_by_data)
```

```python
# Graphical Univariate Analysis
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(df['Session_Duration'], kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Session_duration Distribution')

sns.boxplot(x=df['Session_Duration'], ax=axs[0, 1])
axs[0, 1].set_title('Session_duration Plot')

sns.histplot(df['Total_DL'] + df['Total_UL'], kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Total Data (DL+UL) Distribution')

sns.boxplot(x=df['Total_DL'] + df['Total_UL'], ax=axs[1, 1])
axs[1, 1].set_title('Total Data (DL+UL) Box Plot')

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
applications = ['Social_Media', 'Google', 'Email', 'Youtube', 'Netflix', 'Gaming', 'Other']

for app in applications:
    df[f'{app}_Total'] = df[f'{app}_DL'] + df[f'{app}_UL']

fig, axs = plt.subplots(len(applications), 1, figsize=(10, 20))

for i, app in enumerate(applications):
    sns.scatterplot(x=df[f'{app}_Total'], y=df['Total_DL'] + df['Total_UL'], ax=axs[i])
    axs[i].set_title(f'{app} Data vs. Total Data')
    axs[i].set_xlabel(f'{app} Data (Bytes)')
    axs[i].set_ylabel('Total Data (Bytes)')

plt.tight_layout()
plt.show()
```

```python
# Variable Transformations: Segment users into top five decile classes based on total duration
df['Decile'] = pd.qcut(df['Session_Duration'], 10, labels=False)

decile_data = df.groupby('Decile').agg({'Total_DL': 'sum', 'Total_UL': 'sum'})
decile_data['Total_Data'] = decile_data['Total_DL'] + decile_data['Total_UL']
print(decile_data)
```

```python
from sklearn.decomposition import PCAprin
from sklearn.preprocessing import StandardScaler
```

```python
#Dimentionality Reduction: Principal component Analysis (PCA)
features =  df[[ 'Social_Media_DL', 'Social_Media_UL', 'Google_DL', 'Google_UL',
       'Email_DL', 'Email_UL', 'Youtube_DL', 'Youtube_UL', 'Netflix_DL',
       'Netflix_UL', 'Gaming_DL', 'Gaming_UL', 'Other_DL', 'Other_UL',
       'Total_UL', 'Total_DL']]
# standrdize the features     
features = StandardScaler().fit_transform(features)
#perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(features)
#Creat the Dtaframe with the principal components
pca_df = pd.DataFrame(data = principal_components, columns=['principal_components1','principal_components2'])
```

```python
pca_df
```

```python
#plot the pca reasult
plt.figure(figsize=(10,6))
sns.scatterplot(data= pca_df, x='principal_components1', y='principal_components2')
plt.title('PCA of Principal data' )
plt.show()

#print the explained varience ratio
print('Explained variance ratio:', pca.explained_variance_ratio_)
```
