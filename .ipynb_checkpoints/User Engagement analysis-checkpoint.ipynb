```python
#import libraries
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv('Telecom_challenge_dataset.csv')
```

```python
# Add a total traffic column 
df['total_traffic'] = df['Total_DL'] + df['Total_UL']
```

```python
# Aggregate engagement metrics per customer id (MSISDN/Number)
user_engagement = df.groupby('MSISDN/Number').agg({
    'Number_of_xDR_Sessions':'sum',
    'Session_Duration':'sum', 
    'Social_Media_DL':'sum',
    'Social_Media_UL':'sum', 
    'Google_DL':'sum',
    'Google_UL':'sum',
    'Email_DL':'sum',
    'Email_UL':'sum', 
    'Youtube_DL':'sum',
    'Youtube_UL':'sum', 
    'Netflix_DL':'sum',
    'Netflix_UL':'sum',
    'Gaming_DL':'sum',
    'Gaming_UL':'sum',
    'Other_DL':'sum',
    'Other_UL':'sum',
    'Total_UL':'sum',
    'Total_DL':'sum',
    'Social_Media_Total':'sum',
    'Google_Total':'sum',
    'Email_Total':'sum', 
    'Youtube_Total':'sum',
    'Netflix_Total':'sum',
    'Gaming_Total':'sum',
    'Other_Total':'sum',
}).reset_index
```

```python
# Reporting the top 10 customers per engagement metric
top_10_frequency = user_engagement.nlargest(10, 'Number_of_xDR_Sessions')
top_10_duration = user_engagement.nlargest(10, 'session_duration')
top_10_traffic = user_engagement.nlargest(10, 'total_traffic')

print("Top 10 customers by session frequency:\n", top_10_frequency)
print("Top 10 customers by session duration:\n", top_10_duration)
print("Top 10 customers by total traffic:\n", top_10_traffic)
```
